```python
import thinkplot
```

```python
import nsfg
df = nsfg.ReadFemPreg()
df
```

```python
df.ageqtnur.dropna()
```

```python
df.birthwgt_lb.dropna()
```

```python
birthwgt_lb = []
for i in df.ageqtnur.dropna().keys():
    birthwgt_lb.append(df.birthwgt_lb.dropna().get(i))
birthwgt_lb
```

```python
thinkplot.Scatter(xs=birthwgt_lb, ys=df.ageqtnur.dropna())
thinkplot.Config(xlabel = "Birth Weight (lbs)",
                 ylabel = "Ageqtnur",
                 legend = False)
```
